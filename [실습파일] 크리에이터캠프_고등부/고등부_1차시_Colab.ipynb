# 2. 노트북 인터페이스 익히기

## 2-1. 코드 셀 및 텍스트 셀 사용법
Colab 노트북은 **코드 셀**과 **텍스트 셀**로 구성됩니다.

* **코드 셀 (Code Cell)**
    * Python 코드를 작성하고 실행하는 공간입니다.
    * 셀 왼쪽의 **재생(▶) 버튼**을 누르거나, 단축키(`Shift + Enter`)를 사용하여 실행합니다.

* **텍스트 셀 (Text Cell)**
    * **마크다운(Markdown)** 문법을 사용하여 서식 있는 텍스트를 작성하는 공간입니다.
    * 설명, 제목, 이미지, 링크 등을 추가하여 노트북의 가독성을 높입니다.
    * 셀을 더블클릭하면 편집 모드로 전환됩니다.

```python
# 코드 셀
# 간단한 Python 코드를 작성하고 실행
a = 10
b = 20
print(f"a + b = {a + b}")
print("Shift + Enter 키를 눌러 셀을 실행할 수 있습니다.")
```

## 2-2. 셀 실행 및 순서 관리
* **셀 실행**: 셀을 실행하면 코드의 결과가 셀 바로 아래에 표시됩니다.
* **실행 순서**: 셀은 위에서 아래 순서대로 실행하는 것이 일반적입니다. 이전 셀에서 선언한 변수나 함수는 다음 셀에서 사용할 수 있습니다.
* **주의**: 셀을 순서대로 실행하지 않으면 변수가 정의되지 않았다는 등의 오류가 발생할 수 있습니다. 만약 노트북 상태가 꼬였다면, 상단 메뉴의 **[런타임] > [런타임 다시 시작]**을 통해 초기화할 수 있습니다.

```python
# 이전 코드 셀에서 선언한 변수 a와 b를 사용
# 만약 이전 셀을 실행하지 않고 이 셀을 먼저 실행하면 'a'가 정의되지 않았다는 NameError가 발생
c = a * b
print(f"이전 셀의 변수를 사용하여 계산한 결과: a * b = {c}")
```

## 2-3. 유용한 단축키 소개
Colab에서는 작업 효율을 높여주는 다양한 단축키를 제공합니다.  
단축키는 **명령 모드**(셀을 클릭하여 파란색 테두리가 표시된 상태)에서 동작합니다.

#### **자주 사용하는 필수 단축키**

* **셀 실행**
    * `Shift + Enter`: 현재 셀 실행 후, 다음 셀로 이동
    * `Ctrl + Enter`: 현재 셀 실행 후, 제자리에 머무름
    * `Alt + Enter`: 현재 셀 실행 후, 바로 아래에 새 코드 셀 삽입

* **셀 관리 (명령 모드에서)**
    * `Ctrl + M + A`: 현재 셀 **위에(Above)** 새 셀 추가
    * `Ctrl + M + B`: 현재 셀 **아래에(Below)** 새 셀 추가
    * `Ctrl + M + D`, `D`: 현재 셀 삭제 (D를 두 번 누름)
    * `Ctrl + M + M`: 현재 셀을 **마크다운(Markdown)** 셀로 변경
    * `Ctrl + M + Y`: 현재 셀을 **코드(Code)** 셀로 변경
    * `Ctrl + M + Z`: 셀 삭제 취소

> **팁**: 전체 단축키 목록은 `Ctrl + M H`를 눌러 확인할 수 있습니다.
