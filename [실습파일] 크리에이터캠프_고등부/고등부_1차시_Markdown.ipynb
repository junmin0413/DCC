# Markdown 이란?

- Markdown은 텍스트 기반의 마크업 언어로, 2004년 존 그루버(John Gruber)에 의해 만들어졌습니다.
- 쉽게 쓰고 읽을 수 있으며, HTML로 쉽게 변환할 수 있습니다.
- 특히 Github에서 문서 가독성을 높이는 데 널리 사용되면서 주목받기 시작했습니다.

### **마크다운의 장점**
1.  문법이 간결합니다.
2.  별도의 도구 없이 텍스트 편집기만으로 작성할 수 있습니다.
3.  HTML 등 다양한 형태로 변환이 가능합니다.
4.  텍스트 파일로 저장되어 용량이 매우 작고 보관이 용이합니다.
5.  텍스트 기반이므로 버전 관리 시스템(Git 등)으로 변경 이력을 쉽게 관리할 수 있습니다.
6.  대부분의 플랫폼과 프로그램에서 지원합니다.

### **마크다운의 단점**
1.  공식적인 표준이 없습니다.
2.  표준이 없기 때문에, 사용하는 도구나 플랫폼에 따라 변환 방식이나 결과물이 다를 수 있습니다.
3.  HTML의 모든 기능을 대체하지는 못합니다.

## 1. 제목 (Headers)
'#' 기호를 사용해 제목을 표현합니다. '#'의 개수에 따라 `<h1>`부터 `<h6>`까지 다른 크기의 제목을 만들 수 있습니다.

---

### **✍️ 작성 예시**
# 제목 1 (h1)
## 제목 2 (h2)
### 제목 3 (h3)
#### 제목 4 (h4)
##### 제목 5 (h5)
###### 제목 6 (h6)

제목 (h1 대체)
=====
부제목 (h2 대체)
-----

## 2. 목록 (Lists)
순서가 있는 목록과 순서가 없는 목록을 만들 수 있습니다.

---

* **순서 없는 목록**: `*`, `+`, `-` 기호를 사용합니다.
* **순서 있는 목록**: `1.`, `2.` 와 같이 숫자를 사용합니다.
* **들여쓰기**: `Tab` 키나 공백을 이용해 하위 목록을 만들 수 있습니다.

### **✍️ 작성 예시**
* 첫 번째 항목
  + 두 번째 항목 (들여쓰기)
    - 세 번째 항목 (들여쓰기)

1. 순서 있는 첫 번째 항목
2. 순서 있는 두 번째 항목

## 3. 강조 (Emphasis)
텍스트를 다양하게 강조할 수 있습니다.

---
### **✍️ 작성 예시**

*이탤릭체* 또는 _이탤릭체_

**볼드체** 또는 __볼드체__

~~취소선~~

`인라인(inline) 코드 강조`

## 4. 인용 (Blockquotes)
'>' 기호를 사용하여 인용문을 만듭니다. 중첩하여 사용할 수도 있습니다.

---
### **✍️ 작성 예시**

> 이것은 인용문입니다.
>
> > 이것은 중첩된 인용문입니다.
> >
> > > 이것은 세 번 중첩된 인용문입니다.

## 5. 코드 블록 (Code Blocks)
백틱(```) 세 개로 감싸서 코드 블록을 만듭니다. 시작 부분에 언어 이름을 명시하면 해당 언어의 문법 강조(Syntax Highlighting)가 적용됩니다.

---
### **✍️ 작성 예시**

```python
def say_hello(name):
  print(f"Hello, {name}!")

say_hello("Markdown")
```

## 6. 수평선 (Horizontal Rules)
`---`, `***`, `___` 등 하이픈이나 별표를 3개 이상 연속으로 입력하여 문단을 구분하는 수평선을 만들 수 있습니다.

---
### **✍️ 작성 예시**

---
***
___

## 7. 링크 (Links)
`[보여줄 텍스트](링크 주소 "링크에 마우스를 올렸을 때 나타낼 설명")` 형식으로 링크를 만듭니다.

---
### **✍️ 작성 예시**

**기본 링크:**
[네이버로 이동](https://www.naver.com)

**설명이 있는 링크:**
[구글로 이동](https://www.google.com "Google 검색 페이지")

## 8. 이미지 (Images)

`![대체 텍스트](이미지 경로 또는 URL)` 형식으로 이미지를 삽입합니다. 대체 텍스트는 이미지가 로드되지 않았을 때 표시되는 글입니다.

HTML `<img>` 태그를 사용하면 크기나 정렬 등 더 세부적인 조절이 가능합니다.

---
### **✍️ 작성 예시**

**기본 마크다운 이미지 삽입**

![Google Logo](https://www.google.com/images/branding/googlelogo/2x/googlelogo_color_272x92dp.png)

**HTML 태그를 이용한 크기 및 정렬 조절**

<img src="https://www.google.com/images/branding/googlelogo/2x/googlelogo_color_272x92dp.png" width="200" title="구글 로고">

## 9. 테이블 (Tables)
`|` (파이프)와 `-` (하이픈)을 이용해 표를 만듭니다. 두 번째 줄에서 콜론(`:`)을 사용해 각 열의 텍스트 정렬 방식을 지정할 수 있습니다.

* `:---` : 왼쪽 정렬
* `:---:` : 가운데 정렬
* `---:` : 오른쪽 정렬

---
### **✍️ 작성 예시**

| 제목 1 (왼쪽 정렬) | 제목 2 (가운데 정렬) | 제목 3 (오른쪽 정렬) |
| :--- | :---: | ---: |
| 내용 1 | 내용 2 | 내용 3 |
| 셀 1-1 | 셀 1-2 | 셀 1-3 |

## 10. 수식 (Math Equations)
LaTex 문법을 사용하여 수학 수식을 깔끔하게 표현할 수 있습니다.
* **인라인 수식**: `$` 기호 한 쌍으로 수식을 감싸 문장 안에 포함합니다.
* **블록 수식**: `$$` 기호 두 쌍으로 수식을 감싸 별도의 줄에 가운데 정렬로 표시합니다.

---
### **✍️ 작성 예시**

### 인라인 수식
피타고라스 정리는 문장 안에서 $a^2 + b^2 = c^2$ 와 같이 표현됩니다.

### 블록 수식
근의 공식은 다음과 같이 별도의 단락으로 표시됩니다.
$$x = {-b \pm \sqrt{b^2-4ac} \over 2a}$$

## 11. 텍스트 색상
마크다운 표준 문법은 아니지만, HTML `<font>` 태그를 이용해 텍스트 색상을 변경할 수 있습니다.

---
### **✍️ 작성 예시**

<font color="blue">파란색 텍스트</font> 입니다.  
<font color="#CC3D3D">붉은 계열 텍스트</font> 입니다.
